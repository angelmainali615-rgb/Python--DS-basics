## 2. Looping Statements

Loops are used to execute a block of code repeatedly.

### 2.1 `for` Loops

`for` loops are used to iterate over a *sequence* (like a list, tuple, dictionary, set, or string) or other *iterable* object.

**Syntax:**

```python
for variable in iterable:
    # Code block to execute for each item in the iterable
    # 'variable' takes the value of the current item in each iteration
    print(variable)

````python
user_input = ""
print("Enter 'quit' to exit.")


while user_input != "QUIT":
    user_input = input("Enter command: ")
    print(f"You entered: {user_input}")

print("Exited the loop because input was 'quit'.")

````

````python
count = 0
print("Counting up to 5 using while loop:")
while count < 5:
    print(f"Count is: {count}")
    count +=1

print("Loop finished.")
print("-" * 20)
````

````python
count = 20
print("Counting up to 5 using while loop:")
while count > 5:
    print(f"Count is: {count}")
    count -= 1

print("Loop finished.")
print("-" * 20)
````

````python
# Example 1: break
print("Finding the first multiple of 7 between 1 and 20:")
lists = []
for number in range(1, 21):
   
    if number % 7 == 0:
        print(f"Found it! {number} is a multiple of 7.")
        break
    else:
        lists.append(number)

print("Loop finished or broken.")
print("Lists:", lists)
````

````python
# Example 2: Using 'continue' to print only even numbers
print("Printing even numbers between 1 and 10:")
for number in range(1, 11):
    if number % 2 != 0:
        continue

    print(f"Even number: {number}")

print('Loop finished\'s.')
````

````python
# Example 2: Using 'continue' to print only even numbers
print("Printing even numbers between 1 and 10:")
for number in range(1, 11):
    if number % 2 != 0:
        print("Odd number")

    print(f"Even number: {number}")

print('Loop finished\'s.')
````

````python
print("Using pass when we encounter the number 3:")
for number in range(10):
    if number == 3:
        print("3 is passed")
        pass
    else:
        del number
print("pass example finished")
````
