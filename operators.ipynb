```python
a=10
b=20
print(f"a={a}, b={b}")
print(f"Addition(a+b)={a+b}") #Addition
print(f"Subtraction(a-b)={a-b}") #Subtraction
print(f"multiplication(a*b)={a*b}") 
#Multiplication
```

```python
print(f"Division(a/b)={a/b}")
print(f"Floor Division(a//b)={a//b}")
print(f"Modulus(a%b)={a%b}")
print(f"Exponentiation(a**b)={a**b}")
```

```python
#Comparison Operator
# Comparison operators
x = 10
y = 5

print(f"x = {x}, y = {y}")
print(f"Equal to (x == y): {x == y}")
print(f"Not equal to (x != y): {x != y}")
print(f"Greator than(x>y)={x>y}")
print(f"Less than(x<y)={x<y}")
print(f"Greater tahn or equal to(x>=y)={x>=y}")
print(f"less than or equal to(x<=y)={x<=y}")
```
