```python
def operations(a,b):
    result=a+b
    result1=a-b
    result2=a*b
    result3=a/b

    return result,result1,result2,result3
```

```python
add_result,sub_result,mul_result,div_result=operations(10,7)
print(add_result)
print(sub_result)
print(mul_result)
print(div_result)
```

```python
#*args and **kargs 
# Use *args to pass multiple positional arguments 

def show_args(*args):
    lists=""
    for item in args:
        print(item)
        lists+=item
    return lists

value=show_args("1","2")
print(value)
    
```

```python
def show_args(*args):
    global lists 
    lists=0
    for item in args:
        print(item)
        lists+=int(item)
    return lists

value=show_args("12","13")
print(value)
print(lists)
    
```

```python
sum = 0
def addition(*args):
    for item in args:
        global sum
        sum = sum + item
    return sum

addition(1,2,4,22,44,55,55)
```
