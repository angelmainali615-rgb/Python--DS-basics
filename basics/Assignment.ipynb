```python
import random #to generate a random numbetr

# Computer chooses a random number
secret_number = random.randint(1, 100)

print("Welcome to the Number Guessing Game!")
print("Guess a number between 1 and 100")
guess=0

while guess!=secret_number: #to repeat guess until correct
    guess = int(input("Enter your guess: ")) #to take user input

    if guess > secret_number: #to compare guesses and hints 
        print("Too high")
    elif guess < secret_number:
        print("Too low")
    else:
        print("Congratulations! You guessed the correct number.")
        break
```

Task 1: The
"Cyber-ID" Generator

Scenario: You are building a registration system
for a futuristic gaming convention.

Instructions:

1. Ask the user for
their username, birth_year (as integer),
and is_member (True/False).

2. Calculate their
current age (assume current year is 2025).

3. Create a
variable entry_fee. If they are a member, fee is 10.5,
otherwise 20.0.

4. Output: Print
a formatted ID card:

`"User: [Name]




```python
username=input("Enter your name:")
birth_year=int(input("Enter your birth year:"))
is_member=(input("Are you a member?" )).lower()=="yes"

#calculate age
current_year=2025
age=current_year-birth_year

#set entry fee
if is_member:
    entry_fee=10.5
else:
    entry_fee=20.0

#Print formatted ID card
print("\n--- Gaming Convention ID Card ---")
print(f"User: {username}")
print(f"Age: {age}")
print(f"Member: {is_member}")
print(f"Entry Fee: ${entry_fee}")
```

Task 2: The Smart
Coffee Machine

Scenario: You are programming a vending machine
that checks user credit and sugar preferences.

Instructions:

1. Set a
price: coffee_price = 250.

2. Ask user
for credit_amount and sugar_level (High/Low).

3. Logic:

 - If credit < price: Print
"Insufficient funds."- If credit >= price AND sugar_level is
"High":

Print
"Dispensing Sweet Coffee. Change: [X]"

 - Otherwise:
Print "Dispensing Black Coffee. Change: [X]"


```python
# Smart Coffee Machine

#  Set coffee price
coffee_price = 250

# Take user inputs
credit_amount = int(input("Enter your credit amount: "))
sugar_level = input("Enter sugar level (High/Low): ").lower()

# Apply logic
if credit_amount < coffee_price:
    print("Insufficient funds.")
else:
    change = credit_amount - coffee_price

    if sugar_level == "high":
        print(f"Dispensing Sweet Coffee. Change: {change}")
    else:
        print(f"Dispensing Black Coffee. Change: {change}")
```

```python
ticket_price=300
money=int(input("enter the amount:"))
is_student=input("are you a student? (yes/no)").lower()=="yes"
if is_student:
    final_price=200
else:
    final_price=ticket_price 

if money<final_price:
    
    print("Not enough balance")
else:
    change=money-final_price
    print(f"ticket booked successfully, change:{change}")
    
```
