__List comprehension__

```python
numbers=[1,2,4,5,6,8,9]
squares=[x**2 for x in numbers if x%2==0]
print(squares)
```

```python
y=[]
for x in range(5):
    if x%2==0:
        z=x**2
        y.append(z)

print(y)
```

```python
value=[1,2,3,4,5,6,7,8,9]
square=[x**2 for x in value if x%2==1]
print(square)
```

```python
v=[]
for x in range(1,20):
    if x%2==1:
        y=x**2
        v.append(y)
print(v)
```

```python
s = "Learning python programming"
vowels = [char for char in s if char in "aeiou"]
vowels
```

```python
s = "Learning python programming"
vowels = [char for char in s if char not in "aeiou"]
print(vowels)
```
