### What are Dictionaries?
Unordered collections of key-value pairs

Keys must be unique and immutable (strings, numbers, tuples)

Values can be any Python object

```python
#The pop() method removes the item with the specified key name:

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

```python
for x in thisdict:
  print(x)
```

```python
#Make a copy of a dictionary with the copy() method:

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = thisdict.copy()
print(mydict)
```

```python
#A dictionary can contain dictionaries, this is called nested dictionaries.
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
```

```python
print(myfamily)
```

```python
#update(other_dict): Adds items from other_dict (or iterable) to the dictionary
d1 = {'a': 1}
d1.update({'b': 2, 'c': 3})
print(d1) 
```

```python
#pop(key): Removes the specified key and returns its value
d1.pop('b')
```

```python
print(d1)
```

```python
#popitem(): Removes and returns the last inserted (key, value) pair (LIFO)
print(d1.popitem())
```

```python
#setdefault(key, value): Returns value if key exists; inserts key with value if not.
d1.setdefault('c', 3) # Inserts 'c': 3
d1.setdefault('a', 99) # Does nothing to 'a'
print(d1) 
```

```python
#fromkeys(sequence, value): Creates a new dictionary with keys from sequence and a default value.
keys = ['x', 'y']
new_d = dict.fromkeys(keys, 0)
print(new_d)
```
