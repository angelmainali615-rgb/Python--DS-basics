```python
my_list = ['apple', 'banana', 'cherry']
```

```python
#append:Adds an item to the end of the list
my_list.append('dragonfruit')
print(my_list)
```

__extend(): Adds items of an iterable (like another list) to the end of the current list.__

```python
fruits = ['elderberry', 'fig']
my_list.extend(fruits)
print(my_list)
# Output: ['apple', 'banana', 'cherry', 'dragonfruit', 'elderberry', 'fig']
```

__pop(): Removes and returns the item at a given index. If no index is specified, it removes and returns the last item.__

```python
removed_item = my_list.pop(0) # Remove the first item ('apple')
print(f"Removed: {removed_item}")
print(my_list)
# Output: Removed: apple
# Output: ['blueberry', 'cherry', 'dragonfruit', 'elderberry', 'fig']
```

__clear(): Removes all items from the list.__

```python
my_list.clear()
print(my_list)

```

__index(): Returns the index of the first matched item.__


```python
index = my_list.index('cherry')
print(f"Index of cherry: {index}")

```

__count(): Returns the number of times a specified item appears in the list.__

```python
count = my_list.count('cherry')
print(f"Count of cherry: {count}")

```

__sort(): Sorts the list in ascending order by default__

```python
my_list.sort()
print(my_list)

```

```python
#You can sort in descending order using 
```

```python
my_list.sort(reverse=True)
print(my_list)
```

__reverse(): Reverses the order of the items in the list in place.__


```python
my_list.reverse()
print(my_list)

```

__copy(): Returns a shallow copy of the list__

```python
list_copy = my_list.copy()
print(list_copy)

```

__insert(): Inserts an item at a specified index__

```python
my_list.insert(1, 'blueberry') # Insert at index 1
print(my_list)
```

__remove(): Removes the first occurrence of a specified value from the list.__

```python
my_list.remove('banana')
print(my_list)

```
