The __.format__ method combines values of other data types, e.g., integers,floats,booleans,lists, etc. with strings. You can use __format__ to construct output messages for display.

```python
cost_of_bag= 1.25
profit=.3
Total_bags=400
#Template for output message
output_template= """If a grocery store sells bags at {} per bag with a profit margin of {} then the total profit is {}."""
print(output_template)
```

```python
#inserting values into the string
total_profit= cost_of_bag*profit*Total_bags
output=output_template.format(cost_of_bag,profit*100,total_profit)
print(output)
```

Notice how the placeholders {} in the output_template string are replaced with the arguments provided to the __.format__ method

```python
#When no start index is provided, the interpreter prints all the values from start up to the end index provided.
animals = ["cat", "dog", "bat", "mouse", "pig", "horse", "donkey", "goat", "cow"]
print(animals[::2])  # using positive indexes
print(animals[-8:-1:2])  # using negative indexes
#Here, we have not provided start and end indexes, which means all the values will be considered. But as we have provided a jump index of 2, only alternate values will be printed.
```

```python
#printing every 3rd consecutive within a given range
animals = ["cat", "dog", "bat", "mouse", "pig", "horse", "donkey", "goat", "cow"]
print(animals[1:8:3])
#Here, the jump index is 3. Hence it prints every 3rd element within the given index.
```

```python
#A nested list simply means a list that contains other lists as items.
# Nested List Example
students = [
    ["Harry", 20],
    ["Sarah", 22],
    ["Bruno", 21]
]

print(students)
```

```python
# Accessing elements inside nested lists
students = [
    ["Harry", 20],
    ["Sarah", 22],
    ["Bruno", 21]
]

print(students[0][0])     # First student's name
print(students[1][1])     # Second student's age
```

```python
# Modifying nested list items
students = [
    ["Harry", 20],
    ["Sarah", 22],
    ["Bruno", 21]
]

students[0][1] = 25       # Updating Harry's age

print(students)
```

```python
# Looping through nested lists
students = [
    ["Harry", 20],
    ["Sarah", 22],
    ["Bruno", 21]
]

for student in students:
    print(student)
```

```python
# Looping through values inside nested lists
students = [
    ["Harry", 20],
    ["Sarah", 22],
    ["Bruno", 21]
]

for student in students:
    for value in student:
        print(value)
```
