```python

import datetime as dt

now = dt.datetime.now()
future_date = now + dt.timedelta(days=10)#Uses timedelta to add or subtract durations (days, hours, etc.).
print(future_date)
```

```python
| Code | Meaning           | Example |
| ---- | ----------------- | ------- |
| `%Y` | Year (4 digits)   | 2026    |
| `%y` | Year (2 digits)   | 26      |
| `%m` | Month (01–12)     | 01      |
| `%d` | Day (01–31)       | 05      |
| `%H` | Hour (00–23)      | 14      |
| `%I` | Hour (01–12)      | 02      |
| `%M` | Minute            | 45      |
| `%S` | Second            | 30      |
| `%p` | AM / PM           | PM      |
| `%A` | Full weekday name | Monday  |
| `%B` | Full month name   | January |
```

Using an alias:
We can also use an alias while importing a module. This way we do not need to write the whole name of the module while calling it.

```python
import math as m

print("Square Root of 36 :", m.sqrt(36))
print("3 to the power of 4 :", m.pow(3,4))
```
